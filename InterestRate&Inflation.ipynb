```python

# Renaming columns for better readability
interest_df.columns = [f"col_{i}" if "Unnamed" in col else col.strip() for i, col in enumerate(interest_df.columns)]

# Identifying the date column name
date_column_name = 'F6 - HOUSING LENDING RATES'  # Adjust this if the actual column name is different

# Converting date column to datetime
interest_df[date_column_name] = pd.to_datetime(interest_df[date_column_name])

# Calculate the average interest rate for each date row, excluding the date column
interest_df['Average Interest Rate'] = interest_df.iloc[:, 1:].mean(axis=1)

# Manually create a DataFrame with years from 2019 to 2023
years = pd.DataFrame({'Year': range(2019, 2024)})

# Calculate the yearly average interest rate
interest_df['Year'] = interest_df[date_column_name].dt.year
yearly_avg_interest_df = interest_df.groupby('Year')['Average Interest Rate'].mean().reset_index()

# Ensure we only have data from 2019 to 2023
yearly_avg_interest_df = yearly_avg_interest_df[yearly_avg_interest_df['Year'].isin(range(2019, 2024))]

# Merge the manually created years DataFrame with the calculated averages to ensure all years are present
yearly_avg_interest_df = pd.merge(years, yearly_avg_interest_df, on='Year', how='left')

# Save the DataFrame to a file
output_file_path = "Yearly_Avg_Interest_Rates_2019_2023_v6.xlsx"
yearly_avg_interest_df.to_excel(output_file_path, index=False)

# Display the new DataFrame
yearly_avg_interest_df

```

```python
import pandas as pd
from pathlib import Path

# Define the path to the Excel file
file_path = Path("Inflation_rate_RBA.xlsx")

# Read the Excel file
df = pd.read_excel(file_path)

# Display the original DataFrame to inspect the rows (optional)
print("Original DataFrame:")
print(df.head(15))

# Define the rows to delete (note that row indices are 0-based)
rows_to_delete = list(range(2, 10))

# Drop the specified rows
df_cleaned = df.drop(rows_to_delete)

# Reset the index of the DataFrame after dropping the rows
df_cleaned = df_cleaned.reset_index(drop=True)

# Display the cleaned DataFrame to verify the result
print("\nCleaned DataFrame:")
print(df_cleaned.head(15))
```

```python
# Read the Excel file, skipping the first 11 rows
df = pd.read_excel(file_path, skiprows=11)

# Extract the necessary columns (A and J), assuming they are the first and tenth columns in the DataFrame
df_extracted = df.iloc[:, [0, 9]].copy()

# Convert the date column (first column) to datetime and then extract the year
df_extracted.iloc[:, 0] = pd.to_datetime(df_extracted.iloc[:, 0]).dt.year

# Rename the columns
df_extracted.columns = ['Year', 'Consumer Price Index']

# Display the first few rows of the extracted DataFrame to verify the result
print(df_extracted.head())

# Save the extracted DataFrame to a new Excel file
df_extracted.to_excel("Extracted_Inflation_Rates.xlsx", index=False)
```

```python
# Filter the data for the years 2013-2023
df_filtered = df_extracted[df_extracted['Year'].between(2013, 2023)]

# Group by year and calculate the average Consumer Price Index for each year
df_yearly_avg = df_filtered.groupby('Year', as_index=False)['Consumer Price Index'].mean()

# Display the new DataFrame with yearly averages
print(df_yearly_avg)

# Save the new DataFrame to a new Excel file
df_yearly_avg.to_excel("Yearly_Average_Inflation_Rates.xlsx", index=False)
```

```python
# Merge the interest rate data and housing price data on the year column
combined_df2 = pd.merge(
    df_yearly_avg[['Year', 'Consumer Price Index']],
    Melb_housing_price[['Year', 'Price']],
    on='Year'
)

# Display the merged DataFrame to verify the result
print(combined_df2)
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Consumer Price Index
ax1.set_xlabel('Year')
ax1.set_ylabel('Consumer Price Index (%)', color='tab:green')
ax1.plot(combined_df2['Year'], combined_df2['Consumer Price Index'], color='tab:green', label='Consumer Price Index')
ax1.tick_params(axis='y', labelcolor='tab:green')

# Create a second y-axis to plot Housing Price
ax2 = ax1.twinx()
ax2.set_ylabel('Housing Price ($)', color='tab:red')
ax2.plot(combined_df2['Year'], combined_df2['Price'], color='tab:red', label='Housing Price')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adding a title
plt.title('Consumer Price Index and Housing Prices 2013-2023')

# Show plot with legends
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

plt.show()
```

```python
import scipy.stats as stats

#Calculate the Pearson correlation coefficient
correlation, p_value = stats.pearsonr(combined_df2['Price'], combined_df2['Consumer Price Index'])

#Test
print(stats.ttest_ind(combined_df2['Price'], combined_df2['Consumer Price Index'], equal_var=False))

print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

Statistical Analysis for Inflation

Pearson correlation
A value of 0.495 suggests a moderate positive linear relationship between the two variables, meaning that as one variable increases, the other tends to increase as well.

pvalue
Since the p-value (0.122) is greater than 0.05, we fail to reject the null hypothesis at the 5% significance level.
This means that there is not enough statistical evidence to conclude that the observed correlation is significantly different from zero.

```python
import pandas as pd
from pathlib import Path

# Load the uploaded data file
file_path = Path("Household debt to income.xlsx")
df = pd.read_excel(file_path)

# Filter the data for the years 2013-2023
df_filtered = df[df['Year'].between(2013, 2023)]

# Ensure the data is sorted by Year
df_sorted = df_filtered.sort_values(by='Year')

# Extract the last entry for each year
last_entries = df_sorted.groupby('Year').last().reset_index()

# Select relevant columns: Year and Household Debt to Income
df_yearly_last = last_entries[['Year', 'Household debt to income']]
df_yearly_last.columns = ['Year', 'Household Debt to Income']

# Display the new DataFrame with yearly last month data
print("\nRatio of Household Debt to Annualised Household Disposable Income:")
print(df_yearly_last)

# Save the new DataFrame to a new Excel file
output_path = Path("Household_debt_to_income_last_month.xlsx")
df_yearly_last.to_excel(output_path, index=False)

```
