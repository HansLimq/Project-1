```python
#Importing the pandas,numpy and matplotlib libaries.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
#Creating the data from the 310104.xlsx sheet Data1.
data_frame = pd.read_excel('310104.xlsx',sheet_name='Data1')
#displaying first 10 rows to view the data format
data_frame.head(n = 10)
```

```python
#Displaying the name of the columns
data_frame.columns
```

```python
#Getting the required columns for further analysis.
time_population_vic_df = data_frame.loc[:,['Unnamed: 0','Estimated Resident Population ;  Persons ;  Victoria ;']]
time_population_vic_df.head(10)
```

```python
#checking the first 9 rows  and it  shows 9 rows doesn't contain the required data. So cleaning dataframe for data from 9th row till last row.
time_population_vic_df = time_population_vic_df.loc[9:,:]
time_population_vic_df
```

```python
#Rename the columns.
renaming_columns = {'Unnamed: 0':'Date','Estimated Resident Population ;  Persons ;  Victoria ;':'Estimated Population Victoria'}
time_population_vic_df.rename(columns=renaming_columns,inplace=True)
time_population_vic_df
```

```python
time_population_vic_df['Date']
```

```python
#change to date
time_population_vic_df['Date'] = pd.to_datetime(time_population_vic_df['Date'],format='%Y-%m-%d')

#change population to integer
time_population_vic_df['Estimated Population Victoria'] = time_population_vic_df['Estimated Population Victoria'].astype(int)
#Filter and get only the population data  from 2013 Dec  to 2023 Dec.
time_population_vic_df
```

```python
#Filter and get only the population data  from 2013 Dec  to 2023 Dec.
time_population_vic_2013_2023_dec = time_population_vic_df[(time_population_vic_df['Date']>= '2013-12-01') & (time_population_vic_df['Date']<= '2023-12-31')]
time_population_vic_2013_2023_dec.to_csv("time_population_vic_2013_2023_dec.csv",index=False)
#Plot graph to check the relation Population VS Years.
plt.title("Australia's Victoria Population from 2013 Dec to 2023 Dec ")
plt.xlabel("Years")
plt.ylabel("Populations (millions)")
plt.plot(time_population_vic_2013_2023_dec['Date'], time_population_vic_2013_2023_dec['Estimated Population Victoria'])
plt.show()
```

```python
import pandas as pd
import scipy.stats as stats
# Assuming the data is already loaded into these DataFrames
Melb_housing_price = pd.DataFrame({
    'Year': [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Price': [470395.08, 504963.10, 562489.15, 615270.62, 695069.46, 707043.42, 695089.00, 749727.21, 892358.18, 939518.09, 912255.43]
})
time_population_vic_df = pd.DataFrame({
    'Year': [2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    'Estimated Population Victoria': [5896190, 5998700, 6102230, 6206910, 6312880, 6419890, 6527960, 6637950, 6748970, 6860980, 6974920]
})
# Print column names to verify
print("Melb_housing_price columns:", Melb_housing_price.columns)
print("time_population_vic_df columns:", time_population_vic_df.columns)
# Calculate the Pearson correlation coefficient
correlation, p_value = stats.pearsonr(Melb_housing_price['Price'], time_population_vic_df['Estimated Population Victoria'])
# Test
print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
# Perform t-test
t_stat, p_val = stats.ttest_ind(Melb_housing_price['Price'], time_population_vic_df['Estimated Population Victoria'], equal_var=False)
print(f"T-statistic: {t_stat}")
print(f"P-value: {p_val}")
```
