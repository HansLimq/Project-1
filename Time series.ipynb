```python
import pandas as pd
from pathlib import Path
import matplotlib.pyplot as plt
```

```python
excel_file = Path("Houses-by-suburb-2013-2023.xlsx")
df_Melb = pd.read_excel(excel_file)



```

```python
df_melb_p=df_Melb.rename(columns={2020.0: 2020,  2021.0: 2021, 2022.0: 2022,2023.0:2023 })
```

```python
df_Melb_reduced = df_melb_p.dropna(how='any',axis=0)
df_Melb_reduced.reset_index(drop=True, inplace=True)
```

```python
df_Melb_price = df_Melb_reduced.drop(['Prelim      ','   22-23','13-23','PA'],axis=1)
```

```python
df_Melb_price
```

```python
df_Melb_price.drop(columns=['Locality']).mean(axis=0).index
```

```python
df_Melb_price.set_index('Locality')
mean_2013 = round(df_Melb_price[2013].astype("float").mean(), 2)
mean_2014=round(df_Melb_price[2014].astype("float").mean(), 2)
mean_2015=round(df_Melb_price[2015].astype("float").mean(), 2)
mean_2016=round(df_Melb_price[2016].astype("float").mean(), 2)
mean_2017=round(df_Melb_price[2017].astype("float").mean(), 2)
mean_2018=round(df_Melb_price[2018].astype("float").mean(), 2)
mean_2019=round(df_Melb_price[2019].astype("float").mean(), 2)
mean_2020=round(df_Melb_price[2020].astype("float").mean(), 2)
mean_2021=round(df_Melb_price[2021].astype("float").mean(), 2)
mean_2022=round(df_Melb_price[2022].astype("float").mean(), 2)
mean_2023=round(df_Melb_price[2023].astype("float").mean(), 2)
```

```python
Melb_price_summary = pd.DataFrame({
    2013: [mean_2013],
    2014: [mean_2014],
    2015: [mean_2015],
    2016: [mean_2016],
    2017: [mean_2017],
    2018: [mean_2018],
    2019: [mean_2019],
    2020: [mean_2020],
    2021: [mean_2021],
    2022: [mean_2022],
    2023: [mean_2023],
    })
Melb_price_summary
Melb_housing_price = pd.DataFrame(Melb_price_summary).T
#new_header_1 = Melb_housing_price .iloc[0]
#Melb_housing_price.rename(columns={"0": "Price" })
Melb_housing_price.index
```

```python
df_Melb_price_time_chart=Melb_housing_price.plot(kind="line")
#df_Melb_price_time_chart.set_xlabel("Year")            
#df_Melb_price_time_chart.set_ylabel("Melbourne Hosing Price")
plt.xticks ([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020,2021,2022,2023])
#plt.xticks(list(map(int, Melb_housing_price.index.tolist())))
# Set a title for the chart
plt.title("Melbourne Hosing Price vs. Year")

plt.tight_layout()
plt.show()
```
