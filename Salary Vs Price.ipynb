```python
df_Melb_price = df_Melb_reduced.drop(['Prelim      ','   22-23','13-23','PA'],axis=1)
```

```python
df_Melb_price
```

```python
df_Melb_price.set_index('Locality')
mean_2013 = round(df_Melb_price[2013].astype("float").mean(), 2)
mean_2014=round(df_Melb_price[2014].astype("float").mean(), 2)
mean_2015=round(df_Melb_price[2015].astype("float").mean(), 2)
mean_2016=round(df_Melb_price[2016].astype("float").mean(), 2)
mean_2017=round(df_Melb_price[2017].astype("float").mean(), 2)
mean_2018=round(df_Melb_price[2018].astype("float").mean(), 2)
mean_2019=round(df_Melb_price[2019].astype("float").mean(), 2)
mean_2020=round(df_Melb_price[2020].astype("float").mean(), 2)
mean_2021=round(df_Melb_price[2021].astype("float").mean(), 2)
mean_2022=round(df_Melb_price[2022].astype("float").mean(), 2)
mean_2023=round(df_Melb_price[2023].astype("float").mean(), 2)
```

```python
Melb_price_summary = pd.DataFrame({
    2013: [mean_2013],
    2014: [mean_2014],
    2015: [mean_2015],
    2016: [mean_2016],
    2017: [mean_2017],
    2018: [mean_2018],
    2019: [mean_2019],
    2020: [mean_2020],
    2021: [mean_2021],
    2022: [mean_2022],
    2023: [mean_2023],
    })
Melb_price_summary
Melb_housing_price = pd.DataFrame(Melb_price_summary).T
Melb_housing_price.reset_index(inplace=True)
Melb_housing_price.columns = ['Year', 'Price']
Melb_housing_price
```

```python
df_Melb_price_time_chart=Melb_housing_price.plot(x="Year", y="Price",kind="line")
#df_Melb_price_time_chart.set_xlabel("Year")            
#df_Melb_price_time_chart.set_ylabel("Melbourne Hosing Price")
plt.xticks ([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020,2021,2022,2023])
#plt.xticks(list(map(int, Melb_housing_price.index.tolist())))
# Set a title for the chart
plt.title("Melbourne Hosing Price vs. Year")

plt.tight_layout()
plt.show()
```

```python
Year=Melb_housing_price['Year']
Price=Melb_housing_price['Price']
slope, intercept, rvalue, pvalue, stderr = st.linregress(Year,Price)
p_fit = slope * Year+ intercept
line_price = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_price)
plt.scatter(Year,Price)
plt.plot(Year,p_fit,"-",color="red")
plt.annotate(line_price,(20,15),fontsize=15,color="red")
plt.xticks ([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020,2021,2022,2023])
plt.ylabel('Melbourne Housing Price')
```

```python
fitlter_salary=df_salary_ready[df_salary_ready['Year'].dt.month==11]
```

```python
fitlter_salary['Annual_Salary']= fitlter_salary[' Weekly Income'] * 52
```

```python
fitlter_salary
```

```python
fitlter_salary['Year'] = fitlter_salary['Year'].dt.year
```

```python
fitlter_salary
```

```python
annual_salary_df=fitlter_salary.drop(columns=[' Weekly Income'])
```

```python
annual_salary=annual_salary_df.drop(1)
```

```python
annual_salary.set_index('Year')
```

```python
Year=annual_salary['Year']
Annual_Salary=annual_salary['Annual_Salary']
slope, intercept, rvalue, pvalue, stderr = st.linregress(Year,Annual_Salary)
p_fit = slope * Year+ intercept
line_salary = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_salary)
plt.scatter(Year,Annual_Salary)
plt.plot(Year,p_fit,"-",color="red")
plt.annotate(line_salary,(20,15),fontsize=15,color="red")
plt.xticks ([2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020,2021,2022,2023])
plt.ylabel('Annual_Salary')
```

```python
Salary_price=pd.merge(annual_salary,Melb_housing_price, on=['Year'],how='inner',suffixes=['_and'])
```

```python
Salary_price
```

```python
salary_vs_price_chart= Salary_price.plot(x="Annual_Salary", y="Price", kind="scatter")
salary_vs_price_chart.set_xlabel("Annual_Salary")
salary_vs_price_chart.set_ylabel("Melbourne Housing Price")

# Set a title for the chart
plt.title("Annual_Salary vs. Melbourne House Price")

plt.tight_layout()
plt.show()
```

```python
Annual_Salary=Salary_price['Annual_Salary']
Melbourne_Housing_Price=Salary_price['Price']
slope, intercept, rvalue, pvalue, stderr = st.linregress(Annual_Salary, Melbourne_Housing_Price)
p_fit = slope * Annual_Salary + intercept
line_tem = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
print(line_tem)
plt.scatter(Annual_Salary,Melbourne_Housing_Price)
plt.plot(Annual_Salary,p_fit,"-",color="red")
plt.annotate(line_tem,(20,15),fontsize=15,color="red")
#plt.xticks(TemperatuAnnualMelbourne Housing Pricey')
plt.ylabel('Melbourne Housing Price)')
plt.show()
```

```python
# Calculate the Pearson correlation coefficient
correlation, p_value = st.pearsonr(Salary_price['Price'], Salary_price['Annual_Salary'])
print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
print(f"Pearson correlation coefficient: {correlation}")
print(f"P-value: {p_value}")
```

```python

# Plotting both Average Interest Rate and Housing Price
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Average Interest Rate
ax1.set_xlabel('Year')
ax1.set_ylabel('Salary', color='tab:blue')
ax1.plot(Salary_price['Year'], Salary_price['Annual_Salary'], color='tab:blue', label='Annual Salary')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis to plot Housing Price
ax2 = ax1.twinx()
ax2.set_ylabel('Housing Price ($)', color='tab:red')
ax2.plot(Salary_price['Year'], Salary_price['Price'], color='tab:red', label='Housing Price')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Adding a title
plt.title('Annual Salary and Housing Prices Over Time')

# Show plot with legends
fig.tight_layout()
fig.legend(loc='upper left', bbox_to_anchor=(0.1,0.9))

plt.show()
```
